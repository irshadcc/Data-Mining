

## Data Cleaning, Processing


```python
def get_data(filename,class_column,ratio):
    """
        Returns train_df , val_df, test_df , dict_map
        filename     : Relative filename of the directory
        class_column : Name of the class column
        ratio        : train_ratio
    """
    
    df = pd.read_csv(filename,header=None)
    
    train = { 'x' : [] , 'y' : [] }
    test = { 'x' : [] , 'y' : [] }
    
    classes  = df[class_column].unique().tolist()
    dict_map = dict(enumerate(classes))
    
    for index,clss in dict_map.items():
        
        clss_df = df[df[class_column] == clss].sample(frac=1)
        clss_df = clss_df.drop(columns=class_column)
        n_rows = len(clss_df) 
        
        train_idx_end = math.ceil(n_rows*ratio)
        
        train['x'].append(clss_df.iloc[0 : train_idx_end].values.reshape(-1,4))
        test['x'].append(clss_df.iloc[train_idx_end :].values.reshape(-1,4))      
    
        train['y'].append(np.repeat(index,train_idx_end))
        test['y'].append(np.repeat(index,n_rows-train_idx_end))
    
    train['x'] = np.concatenate(train['x'])
    train['y'] = np.concatenate(train['y'])

    test['x']  = np.concatenate(test['x'])
    test['y']  = np.concatenate(test['y'])

    dict_map['features'] = ['sepal_length','sepal_width','petal_length','petal_width']
    dict_map['classes'] = ['Iris Setosa','Iris Versicolor','Iris Virginica']
    
    return train,test,dict_map
    
    
```

```python
train,test,dict_map = get_data("iris.data",4,0.7)
```

## Decision Tree ( Gini Index , Entropy )

```python
dtree_clf_gini = DecisionTreeClassifier(criterion='gini',random_state=0)
dtree_clf_entropy = DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
dtree_clf_gini.fit(train['x'],train['y'])
dtree_clf_entropy.fit(train['x'],train['y'])
```

```python
export_graphviz(dtree_clf_gini,out_file = "Decision_tree(Gini).dot",feature_names = dict_map['features'],class_names = dict_map['classes'],filled = True,rounded = True)
export_graphviz(dtree_clf_entropy,out_file = "Decision_tree(Entropy).dot",feature_names = dict_map['features'],class_names = dict_map['classes'],filled = True,rounded = True)
```

```python
with open("Decision_tree(Gini).dot") as file:
  dot_graph = file.read()
graphviz.Source(dot_graph)
```

```python
with open("Decision_tree(Entropy).dot") as file:
  dot_graph = file.read()
graphviz.Source(dot_graph)
```

```python
score_gini = dtree_clf_gini.score(test['x'],test['y'])
score_entropy = dtree_clf_entropy.score(test['x'],test['y'])

print("Score Decision Tree (Gini) = ",score_gini)
print("Score Decision Tree (Entropy) = ",score_entropy)
```

## Logistics Regression

```python
log_regress_clf = LogisticRegression(random_state=0)
log_regress_clf.fit(train['x'],train['y'])

score_log_regress = log_regress_clf.score(test['x'],test['y'])
print("Score Logistic Regression = ",score_log_regress)
```
